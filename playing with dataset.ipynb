```python
import pandas as pd

df = pd.read_csv('iris.csv')
df.head()
```

```python
df['species'].value_counts()
```

```python
df['species'].unique()
```

```python
iris_without_species = df.drop(columns=['species'])
iris_without_species.head()
```

```python
species = df.drop(['sepal_length','sepal_width','petal_length','petal_width'], axis=1)
species.head()
```

```python
# calculated column

sepalarea = iris_without_species['sepal_length'] * iris_without_species['sepal_width']

sepalarea.head()
```

```python
petalarea = iris_without_species['petal_length'] * iris_without_species['petal_width']

petalarea.head()
```

```python
iris_without_species['petal_area'] = petalarea

iris_without_species.head()
```

```python
iris_without_species['sepal_area'] = sepalarea

iris_without_species.head(50)
```

```python
# filter out parts of dataframes using pandas

# only part of iris_without_species value of sepal_area is > 14

large_flower = iris_without_species[iris_without_species['sepal_area'] > 15]

large_flower.head(10)
```

```python
large_flower.info()
```

```python
large_flower['sepal_area'].value_counts()
```

```python
large_flower16 = iris_without_species[iris_without_species['sepal_area'] == 16]

large_flower16.info()
```

```python
# sorting values by something

sortedlarge_flower = large_flower.sort_values(by='sepal_area', ascending=False)

sortedlarge_flower.head()
```

```python
largestFlower = sortedlarge_flower['sepal_area']
largestFlower.head(1)
```
